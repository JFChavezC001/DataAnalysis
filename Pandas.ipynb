# Pandas

### -Es un paquete que ayuda a trabajar con datos multidimensionales
### -Significa Panes DataSets
### -Podemos manipular tablas de datos de diferentes tipos en distintos DataFrames

```python
#Lo primero que haremos sera cargar pandas
import pandas as pd
```

```python
#listas de Lugares
Guanajuato =['México', 2019, 'Practica de campo','Callejon del beso',True]
SanLuisPotosi=['México',2019,'Practica de campo','Sótano de las golondrinas',True]
QuintanaRoo=['México',2020,'Practica de campo','Laguna de los 7 colores',True]
Edomex=['México',2022,'servicio social','Zumpahuacán', False]
```

```python
Salidas=[Guanajuato,SanLuisPotosi,QuintanaRoo,Edomex]
```

```python
df_salidas=pd.DataFrame(Salidas, columns=['País','Año de visita','Motivos','Lugar Favorito','VisConGeneracion'])
df_salidas
```

```python
#Creando un DF vacio
ExpLaboral=pd.DataFrame(columns=['Año','Colonia','Municipio','actividades realizadas','Tiempo','Remunerado'])
ExpLaboral
```

```python
#agregar datos a un DF vacio
ExpLaboral.loc[len(ExpLaboral)] = {
    'Año': 2022,
    'Colonia': 'Universidad',
    'Municipio': 'Toluca de Lerdo',
    'Actividades realizadas': 'Analisis de información, Elaboración de material cartografico, Busqueda y recopilación de información',
    'Tiempo':'Agosto 2022 - Enero 2023',
    'Remunerado':True
}
```

```python
ExpLaboral
```

## Seleccionar elementos
### podemos seleccionar elementos a traves del método "loc"

```python
df_salidas
```

```python
#Debemos especificar la posición y el nombre de la columna
df_salidas.loc[2,'Lugar Favorito']
```

```python
#Para realizar seleccion por columna colocamos el nombre del df y el nombre de la columna
df_salidas['Motivos']
```

```python
#por el contrario para seleccionar las filas hacemos uso del metodo iloc
df_salidas.iloc[3]
```

```python
#seleccion booleana
df_salidas["Año de visita"]<2022
```

```python
df_salidas[df_salidas["Año de visita"]<2022]
```
