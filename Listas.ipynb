# LISTAS
### Puede guardar diferentes tipos de datos y/u objetos en un solo espacio.
### La estructura usada para la creación de una lista es:
#### nombre_lista=[val1,val2,val3...]
#### continentes=["América", "Europa", "Africa", "Asia", "Oceania", "Antartida"]
#### altaEscolar=[1822407,'José Francisco Chávez Castillo', 5.5, True]


```python
#para crear una lista siempre se agregan los datos entre llaves
#los datos hacen referencia a clave, ubicacion, año de apertuta, Funcionando
oxxo1=[3543,'Mazatlán, Sinaloa',1997,True]
oxxo2=[3544,'Aquismón, San Luis Potosi', 2001, False]
oxxo3=[3545,'Mérida, Yucatán',2020,True]
oxxo4=[3546,'Guanajuato, Guanajuato', 2007, False]
```

```python
oxxo1[0]
```

```python
#Para seleccionar dos elementos, los encapsulamos entre llaves
[oxxo1[2],oxxo1[1],oxxo1[0],oxxo1[3]]
#funciona del mismo modo si usamos listas distintas
#[oxxo1[0],oxxo2[0],oxxo3[0],oxxo4[0]]
```

```python
#Para seleccionar elementos consecutivos usamos :
[oxxo1[0:3],oxxo2[0:3],oxxo3[0:3],oxxo4[0:3]]
```

```python
#Podemos seleccionar los elementos desde atras
[oxxo1[-3],oxxo2[-4],oxxo3[-4],oxxo4[-4]]
```

## Agregar o Eliminar elementos de una lista

```python
oxxo1
```

```python
#para agregar elementos hacemos uso del metodo append
oxxo1.append('JFCC')
oxxo1
```

```python
#Para eliminar un elemento hacemos uso del metodo remove
oxxo1.remove('JFCC')
oxxo1
```
