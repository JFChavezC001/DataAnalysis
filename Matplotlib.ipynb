# Matplotlib
## Es una libreria para la creación de gráficos 2d y 3d para el análisis de datos

```python
# importamos librerias

import numpy as np
import pandas as pd
```

```python
#listas de Lugares
Guanajuato =['México', 2019, 4,'Practica de campo','Callejon del beso',True]
SanLuisPotosi=['México',2019,5,'Practica de campo','Sótano de las golondrinas',True]
QuintanaRoo=['México',2020,7,'Practica de campo','Laguna de los 7 colores',True]
Edomex=['México',2022,1,'servicio social','Zumpahuacán', False]
```

```python
#importamos la libreria
import matplotlib.pyplot as plt
```

```python
#Ejemplo básico
x=[1,2,3,4,5]
y=[1,8,27,64,125]

#llamamos a la libreria e indicamos la variable de posición a cada punto
plt.plot(x,y)
#Para mostrar usamos la función "show"
plt.show()
```

```python
practicasCampo=[Guanajuato,SanLuisPotosi,QuintanaRoo,Edomex]
practicasCampo
```

```python
df_practicas=pd.DataFrame(practicasCampo, columns=['País','Año de visita','dias duracion','Motivos','Lugar Favorito','VisConGeneracion'])
df_practicas
```

```python
#creamos una gráfica a partir de nuestro df
plt.plot(df_practicas["Año de visita"],df_practicas["dias duracion"])
plt.show()
```

```python
#para crear una grafica de puntos es el mismo proceso pero haciendo uso de scatter
plt.scatter(df_practicas["Año de visita"],df_practicas["dias duracion"])
plt.show()
```
